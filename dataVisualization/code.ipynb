### real data

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]
years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]

plt.plot(years, oscar_revenue)


#add details 
plt.title("Oscars movies revenue in each year")
plt.xlabel("Year")
plt.ylabel("revenue (in $M)")
```

```python
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M


plt.plot(years, oscar_revenue , label = "oscar")
plt.plot(non_oscar_years,non_oscar_revenue , label ="non-oscar")


plt.legend()
plt.title("oscar movies vs non-oscar movies")
plt.xlabel("Year")
plt.ylabel("revenue (in $M)")

```
